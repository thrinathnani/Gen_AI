```python
# ==============================================================================
# ASSIGNMENT: GenAI - Prompt Templates using LangChain
# Objective: Build dynamic and reusable prompt systems.
# ==============================================================================

# Install required libraries if you haven't already:
# !pip install langchain langchain-core

from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

print("="*50)
print("Task 1: Replace Hardcoded Prompts")
print("="*50)

# We use PromptTemplate instead of hardcoded f-strings [cite: 115, 120, 122, 123]
task1_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Testing Task 1
print(task1_template.format(topic="Generative AI"))


print("\n" + "="*50)
print("Task 2: Multi-Input Prompt System")
print("="*50)

# Creating a template with 3 variables [cite: 126, 127-132]
task2_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases [cite: 133-137]
test_cases_t2 = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for test in test_cases_t2:
    print(task2_template.format(topic=test["topic"], audience=test["audience"], tone=test["tone"]))


print("\n" + "="*50)
print("Task 3: Prompt Variations Engine")
print("="*50)

# Creating 3 different templates [cite: 138-142]
teaching_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} clearly step by step")
interview_template = PromptTemplate(input_variables=["topic"], template="Ask 3 questions about {topic}")
storytelling_template = PromptTemplate(input_variables=["topic"], template="Explain {topic} as a story")

topic_t3 = "Machine Learning" # [cite: 145]

print("Teaching:", teaching_template.format(topic=topic_t3))
print("Interview:", interview_template.format(topic=topic_t3))
print("Storytelling:", storytelling_template.format(topic=topic_t3))


print("\n" + "="*50)
print("Task 4: ChatPrompt Template System")
print("="*50)

# Using ChatPromptTemplate to define system and user roles [cite: 146-151]
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert {role}. Act accordingly."),
    ("user", "Let's discuss {topic}.")
])

# Testing different roles [cite: 152-158]
roles = ["teacher", "interviewer", "motivator"]
for role in roles:
    messages = chat_template.format_messages(role=role, topic="Neural Networks")
    print(f"\n--- Role: {role.upper()} ---")
    for msg in messages:
        print(f"{msg.type.capitalize()}: {msg.content}")


print("\n" + "="*50)
print("Task 5: Input Validation Layer")
print("="*50)

def validate_inputs(audience, tone):
    """Validates inputs and assigns default values if invalid [cite: 161-168]"""
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Validation logic: Fallback to defaults if not in valid lists
    final_audience = audience.lower() if audience.lower() in valid_audiences else "beginner"
    final_tone = tone.lower() if tone.lower() in valid_tones else "casual"

    if final_audience != audience.lower() or final_tone != tone.lower():
        print(f"Warning: Invalid inputs detected. Defaulting to audience='{final_audience}', tone='{final_tone}'")

    return final_audience, final_tone

# Testing Validation
print("Test 1 (Valid):", validate_inputs("expert", "formal"))
print("Test 2 (Invalid):", validate_inputs("aliens", "angry"))


print("\n" + "="*50)
print("Task 6: Prompt Generator App")
print("="*50)

def generate_prompt(topic, audience, tone, style):
    """App that validates inputs, selects a style, and formats the template [cite: 170-177]"""
    # 1. Validate Inputs
    val_audience, val_tone = validate_inputs(audience, tone)

    # 2. Define style templates (Logic separate from templates) [cite: 207]
    templates = {
        "teaching": PromptTemplate(input_variables=["topic", "audience", "tone"], template="Teach {topic} to {audience} users. Keep the tone {tone} and go step by step."),
        "interview": PromptTemplate(input_variables=["topic", "audience", "tone"], template="Conduct an interview about {topic} suitable for {audience} level candidates. Use a {tone} tone."),
        "storytelling": PromptTemplate(input_variables=["topic", "audience", "tone"], template="Explain {topic} for {audience} in a {tone} storytelling style.")
    }

    # 3. Select template (default to teaching if style is unknown)
    selected_template = templates.get(style.lower(), templates["teaching"])

    # 4. Generate Output
    return selected_template.format(topic=topic, audience=val_audience, tone=val_tone)

# Example Output matching documentation [cite: 176, 177]
print(generate_prompt(topic="Neural Networks", audience="beginners", tone="fun", style="storytelling"))


print("\n" + "="*50)
print("Task 7: Template Reusability Test")
print("="*50)

# Using ONE template with 5 different inputs [cite: 178-184]
reusable_template = PromptTemplate(
    input_variables=["domain", "task"],
    template="Act as a senior {domain} professional. Your task is to {task}."
)

reusability_inputs = [
    {"domain": "Data Science", "task": "explain the difference between random forest and XGBoost"},
    {"domain": "Software Engineering", "task": "write a Python script to reverse a linked list"},
    {"domain": "Cybersecurity", "task": "outline a plan to secure a cloud database"},
    {"domain": "Digital Marketing", "task": "draft a 3-month SEO strategy for a startup"},
    {"domain": "Finance", "task": "summarize the risks of trading options"}
]

for i, test in enumerate(reusability_inputs, 1):
    print(f"Generation {i}:")
    print(reusable_template.format(domain=test["domain"], task=test["task"]))
    print("-" * 30)
```
